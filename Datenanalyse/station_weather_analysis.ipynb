```python
sql('''select s.id, s.name, avg(w.tmin) from weather w
join station s on s.id = w.id
where s.state = 'CA'
and extract(year from w.date) = 2018
group by s.id;''')
```

```python
sql('''select s.state, avg(w.tmax) tmax from weather w
join station s on s.id = w.id
and extract(year from w.date) = 2018
and w.tmax <> 'NaN'
group by s.state
order by tmax;''')
```

```python
tmax = sql('''select s.state, extract(year from w.date), avg(w.tmax) tmax from weather w
join station s on s.id = w.id
where (s.state = 'FL' or s.state = 'HI' or s.state = 'TX' or s.state = 'LA')
and w.tmax <> 'NaN'
group by s.state, extract(year from w.date)
order by extract(year from w.date);''')
```

```python
tmax['tmax'] = tmax['tmax'].apply(lambda x: x/10)
```

```python
tmax = tmax.pivot(index=['date_part'], columns='state', values='tmax')
```

```python
tmax = tmax.reset_index()
```

```python
tmax['date_part'] = tmax['date_part'].apply(int)
```

```python
tmax
```

```python
tmax.plot(x='date_part', y=['FL', 'HI', 'TX', 'LA'], xticks=range(2004,2022), xlabel = 'Jahr', ylabel = 'Durchschnittliche Höchsttemperatur in °C', figsize=(10,5))
```

```python
tavg = sql('''select s.state, extract(year from w.date) as year, extract(month from w.date) as month, avg(((w.tmax)+(w.tmin))/2) tavg from weather w
join station s on s.id = w.id
and w.tmax <> 'NaN'
and w.tmin <> 'NaN'
group by s.state, year, month
order by year, month;''')
tavg
```

```python
tavg_save = tavg
```

```python
tavg['tavg'] = tavg['tavg'].apply(lambda x: x/10)
tavg = tavg.pivot(index=['year', 'month'], columns='state', values='tavg')
tavg = tavg.reset_index()
tavg['year'] = tavg['year'].apply(int)
tavg['month'] = tavg['month'].apply(int)
tavg
```

```python
tavg_time = tavg.set_index(pd.to_datetime(tavg[['year','month']].assign(day=1)))
tavg_time.plot(y=['FL', 'CA'], xlabel = 'Jahr', ylabel = 'Niederschlag', title = 'Durchschnittlicher Niederschlag', figsize=(20,10))
```

```python
prcp = sql('''select s.state, extract(year from w.date) as year, extract(month from w.date) as month, avg(prcp) prcp from weather w
join station s on s.id = w.id
and w.prcp <> 'NaN'
group by s.state, year, month
order by year, month;''')
prcp 
```
